```python
import pandas as pd
```

```python
success = pd.read_table("q_learn.txt", header=None)
params = success[0:len(success):2].reset_index(drop=True)
params.columns = ["Parameters"]
rate = success[1:len(success):2].reset_index(drop=True)
params['Success Rate'] = rate
```

```python
params.sort_values('Success Rate', ascending=False)
```
